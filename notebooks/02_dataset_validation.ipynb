```python
import json
import pandas as pd
from pathlib import Path

# Load your dataset
# data_path = Path("data/arxiv_100k_abstracts.json")
data_path = '../data/arxiv_100k_abstracts.json'

# Quick inspection
with open(data_path, 'r', encoding='utf-8') as f:
    # Check if it's JSON lines or single JSON array
    first_line = f.readline()
    f.seek(0)
    
    if first_line.startswith('['):
        # Single JSON array
        data = json.load(f)
        print(f"Format: Single JSON array")
    else:
        # JSON lines
        data = [json.loads(line) for line in f]
        print(f"Format: JSON lines")

print(f"\nTotal documents: {len(data)}")
print(f"\nSample document keys: {list(data[0].keys())}")
print(f"\nFirst abstract preview:")
print(data[0].get('abstract', 'N/A')[:200])
```
